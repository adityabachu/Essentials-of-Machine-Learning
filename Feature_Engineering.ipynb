# Encoding categorical(ordinal) variables

```python
from sklearn import preprocessing
marks_science = [78, 56, 87, 91,45, 62]
mark_maths = [75, 62, 90, 95, 42, 57]
grade = ['B', 'C', 'A', 'A', 'D', 'B']
data = pd.DataFrame({'Science marks' : marks_science, 'Maths marks' : mark_maths, 'Grade' : grade})
data
```

# Option 1

```python
le = preprocessing.LabelEncoder()
le.fit(data['Grade'])
data['Grade'] = le.transform(data['Grade'])
```

```python
data
```

# Option 2

```python
target = data['Grade'].replace(['A', 'B', 'C', 'D'], [0, 1, 2, 3])
```

```python
predictors = data.iloc[:, 0:2]
data_mod = pd.concat([predictors.reset_index(drop = True), target], axis = 1)
```

```python
data_mod
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
iris = datasets.load_iris()
predictors = iris.data[:, 0:4]
```

```python
iris
```

```python
predictors
```

```python
target = iris.target
predictors = StandardScaler().fit_transform(predictors)# Standardizing the predictors
```

```python
predictors
```

```python
pca = PCA(n_components = 2)
princomp = pca.fit_transform(predictors)
princomp_ds = pd.DataFrame(data = princomp, columns = ['PC 1', 'PC 2'])
princomp_ds
```

```python
princomp
```

```python
target
```

```python
target_ds = pd.DataFrame(data = target, columns = ['class'])
target_ds
```

```python
data_mod = pd.concat([princomp_ds.reset_index(drop = True), target_ds], axis = 1)
fig = plt.figure(figsize = (8,8))
pca_plot = fig.add_subplot(1,1,1)
pca_plot.set_xlabel('PC1')
pca_plot.set_ylabel('PC2')
pca_plot.set_title('PCA (2 component)', fontsize = 20)
classes = [0, 1, 2]
colors = ['y', 'r', 'b']
for target, color in zip(classes, colors):
  indices = data_mod['class'] == target
  pca_plot.scatter(data_mod.loc[indices, 'PC 1'], data_mod.loc[indices, 'PC 2'])
legends = ['Setosa', 'versicolor', 'Virginica']
pca_plot.legend(legends)
```

# Feature subset selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
data = pd.read_csv("apndcts.csv")
predictors = data.iloc[:, 0:7] #Seggregating the predictor variables...
target = data.iloc[:,7] #Segregating the target/ class variable...
test = SelectKBest(chi2, k = 2)
fit = test.fit(predictors, target)
features = fit.transform(predictors)
predictors.columns
```

```python
np.set_printoptions(precision = 3)
print(fit.scores_)
```
