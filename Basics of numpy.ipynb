# Array Concatenation

## create two arrays of same size

```python
a1 = np.array([[1,2,3],[4,5,6]])
```

```python
a2 = np.array([[7,8,9],[10,11,12]])
```

```python
a1
```

```python
a2
```

```python
concat_array = np.concatenate([a1,a2],axis = 0)
```

## axis = 0 is row-wise concatenation

```python
concat_array
```

```python
concat_array = np.concatenate([a1,a2],axis = 1)
```

## axis = 1 is column-wise concatenation

```python
concat_array
```

```python
a1 = np.array([[1,1],[0,1]])
a2 = np.array([[2,0],[3,4]])
```

```python
a1 @ a2
```

```python
a1.dot(a2)
```
